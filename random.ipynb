```python
import json
from sklearn.model_selection import train_test_split

def load_and_flatten_data(file_path):
    flattened_data = []
    
    with open(file_path, 'r') as f:
        for line in f:
            entry = json.loads(line)
            text = entry['Text']
            
            # Iterate through all opinions in the sentence
            for quad in entry['Quadruplet']:
                aspect = quad['Aspect']
                category = quad['Category']
                va_string = quad['VA'] # "6.75#6.38"
                
                # --- LOGIC: Handle NULL Aspect ---
                # If aspect is explicitly named ("soup"), use it.
                # If aspect is "NULL", use the Category ("RESTAURANT#PRICES") as the target.
                if aspect == "NULL":
                    target_text = category.replace("#", " ") # Make "RESTAURANT PRICES" readable
                else:
                    target_text = aspect

                # --- LOGIC: Parse Scores ---
                val_score, aro_score = map(float, va_string.split('#'))
                
                flattened_data.append({
                    "text": text,
                    "target": target_text,
                    "Opinion" :
                    "labels": [val_score, aro_score] # The 2 numbers we want to predict
                })
                
    return flattened_data

# Usage
all_data = load_and_flatten_data('/Users/anatoliifrolov/Desktop/epstein files/extractor/dimabsa/eng_restaurant_train_alltasks.jsonl')

# Split into Train and Validation sets (80% train, 20% val)
train_data, val_data = train_test_split(all_data, test_size=0.2, random_state=42)

print(f"Training items: {len(train_data)}")
print(f"Example Item: {train_data[0]}")
```
